```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate random data set
X_train = np.sort(np.random.uniform(low=1, high=2, size=(50,)))  #features
y_train = np.sort(np.random.uniform(low=300, high=500, size=(50,)))   #target value
```

```python
# Plot the data points
plt.scatter(X_train, y_train, marker='x', c='r')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
# Create and fit the model
linear_model = LinearRegression()
# X must be a 2-D Matrix
linear_model.fit(X_train.reshape(-1, 1), y_train) 
```

```python
# View parameters
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
print(f"'manual' prediction: f_wb = wx+b : {1200*w + b}")
```

```python
# final_f_wb = compute_model_output(x_train, w, b,)
final_f_wb = np.dot(X_train.reshape(-1, 1), w) + b

# Plot our model prediction
plt.plot(X_train, final_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(X_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# Make prediction
X_test = np.array([[1200]])
print(f"Prediction for 1200 sqft house: ${linear_model.predict(X_test)[0]:0.2f}")
```

```python
# load the dataset for multi-feature regression
data = np.loadtxt("./houses.txt", delimiter=',', skiprows=1)
X_train = data[:,:4]
y_train = data[:,4] 
X_features = ['size(sqft)','bedrooms','floors','age']
```

```python
# Fit the model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train) 
```

```python
# View the parameters
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w:}, b = {b:0.2f}")
```

```python
# Predict using the model
print(f"Prediction on training set:\n {linear_model.predict(X_train)[:4]}" )
print(f"prediction using w,b:\n {(X_train @ w + b)[:4]}")
print(f"Target values \n {y_train[:4]}")

x_house = np.array([1200, 3,1, 40]).reshape(-1,4)
x_house_predict = linear_model.predict(x_house)[0]
print(f" predicted price of a house with 1200 sqft, 3 bedrooms, 1 floor, 40 years old = ${x_house_predict*1000:0.2f}")
```
